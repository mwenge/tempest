Create a list of ROMs and their SHA1 signatures.

````python
import re
rom_shas = {}
for line in ROM_DETAIL.split('\n'):
    if "_LOAD" not in line:
        continue
    l = line.split(',')
    rom = re.search('"(.*)"', l[0]).group(1)
    sha = re.search('SHA1\((.*?)\)', l[3]).group(1)
    rom_shas[rom] = sha
````

Check the SHA1 on all our ROMs

````python
import hashlib
import os

for f,s in rom_shas.items():
    if not os.path.isfile("roms/"+f):
        print(f, "missing")
        continue
    d = open("roms/"+f,'rb').read()
    sha = hashlib.sha1(d).hexdigest()
    if s != sha:
        print(f,"different",rom_shas[f], sha)
        continue
````
